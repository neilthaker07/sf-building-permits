```python
use_sf = sf_permits[['Zipcode','Permit Type','Filed Date','Issued Date','Completed Date']]
```

```python
use_sf['end_date']=np.where((use_sf['Completed Date'].isnull()),use_sf['Issued Date'],use_sf['Completed Date'])
```

```python
df_end = use_sf.isnull().sum()
```

```python
df_end
```

```python
df_fields = use_sf[['Zipcode','Permit Type', 'Filed Date', 'end_date']].copy()
```

```python
df_fields.info()
```

```python
df_fields = df_fields.dropna()
```

```python
df_fields.info()
```

```python
df_fields.columns = ['zipcode', 'permit_types','from_date','to_date']
```

```python
df_fields
```

```python
df_fields['from_date']=pd.to_datetime(df_fields['from_date'])
```

```python
df_fields['to_date']=pd.to_datetime(df_fields['to_date'])
```

```python
df_fields.info()
```

```python
df_fields['process_time']=df_fields['to_date']-df_fields['from_date']
```

```python
df_fields
```

```python
df_prep = df_fields[['zipcode','permit_types', 'process_time']].copy()
```

```python
# convert date time days to only number of days to improve the model
```

```python
df_x=df_prep[['zipcode','permit_types']].copy()
df_y=df_prep['process_time']
```

```python
from sklearn import datasets, linear_model
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
reg=linear_model.LinearRegression()
```

```python
x_train,x_test,y_train,y_test = train_test_split(df_x,df_y,test_size=0.2,random_state=4)
```

```python
reg.fit(x_train,y_train)
```

```python
reg.coef_
```

```python
a=reg.predict(x_test)
```
